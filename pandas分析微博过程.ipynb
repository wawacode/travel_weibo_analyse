看一下微博方面的数据data

先分组，分组的目的把相同旅游局的放在一起，计算他发帖的h指数

```python
data.columns
```

根据用户名来分组

```python
gb=data.groupby('用户名')
```

size()是显示分组后的统计信息(发表多少篇微博，索引值)

```python
gb1=gb.size()
```

```python
gbindex=gb1.index
```

```python
gbindex,gb1
```

先产生DataFrame，四个字段,转发的h指数，评论h指数，点赞h指数,几何平均h指数 fh ch lh max_hdd

再产生这个DataFrame之前排序

```python
sortallf=data.sort_values(by=['用户名','转发数'],ascending=[True,False])
```

```python
sortallf
```

```python
sortallc=data.sort_values(by=['用户名','评论数'],ascending=[True,False])
```

```python
sortalll=data.sort_values(by=['用户名','点赞数'],ascending=[True,False])
```

```python
gbindex.size
```

```python
all_h=pandas.DataFrame(numpy.arange(34*4).reshape(34,4),index=gbindex,columns=['fh','ch','lh','max_hdd'])
```

```python
all_h
```

定义个三个数组fh,ch,lh存放转发h指数，评论h指数,点赞h指数,对于每个用户的最大值
循环查找，找到每条微博对应的h指数

```python
fh=[]
ch=[]
lh=[]
max_hdd=[]
mm=(sortallf,sortallc,sortalll)
for j in range(len(mm)):
    for i in gbindex:
        tempdf=mm[j][mm[j]['用户名']==i]
        tempdf['numf']=range(len(tempdf))
        if j==0:
                a=len(tempdf[tempdf['转发数']>=tempdf['numf']+1])
                fh.append(a)
        elif j==1:
                b=len(tempdf[tempdf['评论数']>=tempdf['numf']+1])
                ch.append(b)
        else:
                c=len(tempdf[tempdf['点赞数']>=tempdf['numf']+1])
                lh.append(c)
```

```python
all_h['fh']=fh
```

```python
all_h['ch']=ch
```

```python
all_h['lh']=lh
```

```python
all_h.describe()
```

```python
all_h
```

```python
all_h.drop(['HS1','HS2','HS3','HS4'],axis=1,inplace=True)
```

```python
all_h.drop(['max_hdd'],axis=1,inplace=True)
```

```python
all_h.drop(['HS'],axis=1,inplace=True)
```

```python
all_h
```

```python
all_h.insert(3,'HS',all_h.mean(axis=1))
```
